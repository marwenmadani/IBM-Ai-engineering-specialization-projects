```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength']
```

```python
predictors.head()
```

```python
target.head()
```

```python
#Normalizing iput 
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1] # number of predictors
n_cols
```

# A

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Defining The Model

def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
#Split data into train/test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( predictors, target, test_size=0.3, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# build the model
model = regression_model()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
lmqr=[]

for n in range (1,50):
    model.fit(X_train, y_train, epochs=50)
    pmodel= model.predict(X_test)
    mqr=mean_squared_error(y_test, pmodel)
    lmqr.append(mqr)
    
mean=np.mean(lmqr)
STD=np.std(lmqr)
print(mean)
print(STD)
```

### Mean = 83.205
### Standard deviation = 150.822

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split( predictors_norm, target, test_size=0.3, random_state=4)
print ('Train set:', X_train1.shape,  y_train1.shape)
print ('Test set:', X_test1.shape,  y_test1.shape)
```

```python
lmqr1=[]
model1 = regression_model()
for n in range (1,50):
    model1.fit(X_train1, y_train1, epochs=50)
    pmodel1= model1.predict(X_test1)
    mqr1=mean_squared_error(y_test1, pmodel1)
    lmqr1.append(mqr1)
    
mean1=np.mean(lmqr1)
STD1=np.std(lmqr1)
print(mean1)
print(STD1)
```

### Mean = 70.306
### Standard deviation = 40.340

### ==> Using normalizied Data yeilded in a better performing model
